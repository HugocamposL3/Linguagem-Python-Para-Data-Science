## Lista de Números Aleatórios - Colocar um número que vai ser o tanto de indices que a lista vai ter e printar em ordem aleatória o números que estão na lista.


```python
import random

a = int(input("Qual é o tamanho da Lista "))

Lista = []

while a > 0:
  Lista.append(random.randint(0,100))
  a-=1
print(Lista)
```
