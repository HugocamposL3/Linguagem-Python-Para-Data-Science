```python
# If_simples
aluno = int (input('qual é a idade do aluno'))
if aluno>=12:
  print("Você pode jogar")
if aluno<12:
  print('você não pode jogar')
  
```

```python
# If_composto
doaçao = round(float (input ('Qual o valor da doação: ')),2 )
if doaçao<=1000:
  doaçãoresultado1 = doaçao*0.05
  print ('Valor a ser doado é: {:.2f}'.format(doaçãoresultado1))
else:
  doaçãoresultado2 = doaçao*0.15
  print('valor a ser investido é: {:.2f}'.format(doaçãoresultado2))

```

```python
# If_encadeado
idade = int(input("Qual é a idade do aluno? "))
if (idade<16):
  print("Não pode Votar e nem embacar")
else:
  if idade >= 16 and idade < 18:
    print ('Está a sua escolha')
  else: 
     print ('seu voto é obragtório')
  
```
