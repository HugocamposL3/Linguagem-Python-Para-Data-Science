## Aula que aprendemos as diferenças entre Listas, Tuplas e Dicionário.

## Listas

```python
Top4 = ["Lewis","Ricciardo","Vettel","Lando"]
for nome in Top4:
  print(nome)
Top4.append("Ocon") ## adciona algo no final da lista
for nome in Top4:
  print(nome)

```

```python
Top4 = ["Lewis","Ricciardo","Vettel","Lando"]
for nome in Top4:
  print(nome)
Top4.insert(2,"Ocon") ## adciona algo no lugar da lista que eu estipulei
for nome in Top4:
  print(nome)
```

```python
Top4 = ["Lewis","Ricciardo","Vettel","Lando"]
for nome in Top4:
  print(nome)
Top4.pop() ## deleta o ultimo valor da lista
for nome in Top4:
  print(nome)
```

```python
Top4 = ["Lewis","Ricciardo","Vettel","Lando"]
for nome in Top4:
  print(nome)
Top4.pop(-1) ## deleta o valor da lista que eu estipulei no 'pop'
for nome in Top4:
  print(nome)
```

```python
Top4 = ["Lewis","Ricciardo","Vettel","Lando"]
for nome in Top4:
  print(nome)
Top4.pop(2) ## deleta o valor da lista que eu estipulei no 'pop'
for nome in Top4:
  print(nome)
```

```python
Top4 = ["Lewis","Ricciardo","Vettel","Lando",'Lewis','Lewis','Lewis']
for nome in Top4:
  print(nome)
contagem=Top4.count('Lewis') ## Conta quantos valores especifico tem na lista 
print("Nessa lista o nome Lewis aparece {} vezes".format(contagem))
```

```python
Top4 = ["Lewis","Ricciardo","Vettel","Lando"]
for nome in Top4:
  print(nome)
Top4.sort() ## Ordena os valores na ordem crescentes ou alfabética
for nome in Top4:
  print(nome)
```

```python
Top4 = ["Lewis","Ricciardo","Vettel","Lando"]
for nome in Top4:
  print(nome)
Top4.reverse() ## Eu inverto a ordem de elementos que tenho na lista 
for nome in Top4:
  print(nome)
```

## Dicionário

```python
pilotos = {"Hamilton":"Mercedes","Lando":"Mclaren","Ricciardo":"Mclaren","Max Verstappen":"RBR","Kimmi":"Alfa Romeo"} ## Chave/valor
print(pilotos["Ricciardo"])
```

```python
pilotor = {"Hamilton":"Mercedes","Lando":"Mclaren","Ricciardo":"Mclaren","Max Verstappen":"RBR","Kimmi":"Alfa Romeo"} ## Chave/valor
for chave in pilotos.keys():
  print(chave)
```

```python
pilotos = {"Hamilton":"Mercedes","Lando":"Mclaren","Ricciardo":"Mclaren","Max Verstappen":"RBR","Kimmi":"Alfa Romeo"} ## Chave/valor
for chave,valor in pilotos.items():
  print("O Piloto {} é da equipe {}".format(chave,valor))
```

```python
pilotos = {"Hamilton":"Mercedes","Lando":"Mclaren","Ricciardo":"Mclaren","Max Verstappen":"RBR","Kimmi":"Alfa Romeo"} ## Chave/valor
pilotos["Bottas"]='Mercedes'
for chave,valor in pilotos.items():
  print("O Piloto {} é da equipe {}".format(chave,valor))
```
