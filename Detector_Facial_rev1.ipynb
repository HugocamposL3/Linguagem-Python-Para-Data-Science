## Importando Bibliotecas Necess√°rias

```python
import cv2
from matplotlib import pyplot as plt
```

## Carregando Arquivo XML

```python
carregaAlgoritmo = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

##Carregando a Imagem 

```python
imagem = cv2.imread('/content/sample_data/imagens/imagem6.jpg')
```

##Plotando a Imagem Usando MatplotLib

```python
plt.imshow(imagem) # A imagem sai no modo BGR 
```

##Transformando a Imagem em Modo RGB

```python
image_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

```python
plt.figure()
plt.imshow(image_rgb)
plt.show()
```

##Transformando a Imagem tom de Cinza porque a Acuracia √© melhor

```python
imagemCinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(imagemCinza)
```

## Comando Para Detec√ß√£o de Faces

```python
faces = carregaAlgoritmo.detectMultiScale(image_rgb)
print(faces)
```

```python
for(x, y, l, a) in faces:
    cv2.rectangle(image_rgb, (x, y), (x + l, y + a), (0, 255, 0), 2)
```

```python
plt.imshow(image_rgb)
plt.show()
```
