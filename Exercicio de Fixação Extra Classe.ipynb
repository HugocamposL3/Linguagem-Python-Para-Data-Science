```python
#Faça um programa que leia um número inteiro e mostre na tela o seu sucessor e seu antecessor
print ('Olá! Este é um Programa para advinhar o núemero seguinte e o antecessor')
numero_plot = int (input ('Escreva um número qualquer: '))
valor_ant = numero_plot - 1
valor_suc = numero_plot + 1
print ('Os valores a serem apresentados são esses\nValor Sucessor: {}\nValor Antecessor: {}'.format(valor_suc, valor_ant))

```

```python
# Crie um algoritmo que leia um número e mostra o seu dobro, triplo e raiz quadrada
print ('Olá! Este é um Programa para advinhar o triplo do numero estipulado, dobro e sua raiz quadrada')
import math
numero_plot = int (input ('Escreva um número qualque: '))
valor_dob = numero_plot*2
valor_tri = numero_plot*3
valor_raiz = math.sqrt(numero_plot)
print ('Os valores a serem apresentados são esses\nValor Dobrado: {}\nValor Triplo: {}\nRaiz Quadrada do Valor: {}'.format(valor_dob, valor_tri, valor_raiz))
```

```python
# - Desenvolva um programa que leia as duas notas de um aluno, calcule e mostre a sua média
nota1 = float (input ('Escreva a primeira nota: '))
nota2 = float (input ('Escreva a segunda nota: '))
media = ((nota1 + nota2)/2)
print ('A média do aluno é: {}'.format(media))
```

```python
#  Escreva um programa que leia um valor em metros e o exiba convertido em centímetros e milímetros
valor_metro = float (input('Escreva a distância em metro: '))
conv_cent = valor_metro*100
conv_mili = valor_metro*1000
print ('Distância em Centímetro: {}Cm\nDistância em Milímetro: {}mm'.format(conv_cent,conv_mili))
```

```python
# Crie um programa que leia quanto dinheiro uma pessoa tem na carteira e mostre quantos dólares ela pode comprar - considerar - US$1,00 = R$ 3,27
valor_real = float (input('Escreva quanto você em Real: '))
valor_dolar = valor_real/3.27
print ('Você pode comprar {:.2f} dolares'.format(valor_dolar))
```

```python
# Faça um programa que leia a largura e a altura de uma parede em metros e calcula a sua área e a quantidade de tinta necessária para pinta-la, sabendo que cada litro de tinta, pinta uma área de 2m quadrados
largura = float (input('A largura é: '))
altura = float (input('A altura é: '))
area = altura*largura
tinta = area/2
print ('Tamanho da área é: {:.2f}m²\nE você vai precisar de {} litros de Tinta'.format(area,tinta))
```

```python
# - Faça um algoritmo que leia o preço de um produto e mostre seu novo preço, com 5% de desconto
preço = float(input('Qual é o preço do produto? R$'))
desconto = preço*0.05
precofinal = preço-desconto
print("O preço do produto com Desconto é: R${}" .format(precofinal))
```

```python
#  Faça um algoritmo que leia o salário de um funcionário e mostre seu novo salário, com 15% de aumento.
salario = float (input('Qual é o salario do Colaborador? R$'))
aumento = salario*0.15
salariofinal = salario+aumento
print("O Salário Final é: R${:.2f}".format(salariofinal))
```

```python
# Faça um programa que leia o comprimento do cateto oposto e do cateto adjacente de um triangulo retângulo, calcule e mostre o comprimento da hipotenusa
catopo = float (input('Qual o comprimento do Cateto Oposto? '))
catadj = float (input("Qual o comprimento do Cateto Adjacente? "))
print("A hipotenusa é: ",math.hypot(catopo,catadj))
```
