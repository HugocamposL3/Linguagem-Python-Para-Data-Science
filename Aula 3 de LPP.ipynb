```python
## Programa da Máquina Bimanual
from time import time
botao_1 = int(input('Acione 1 Para Ligado e 0 Para Desligado'))

tempo1 = time()
botao_2 = int(input('Acione 1 Para Ligado e 0 Para Desligado1'))
tempo2 = time()
if(botao_1 == 1 and botao_2 == 1 and (abs(tempo1-tempo2) < 0.5)): ## A função "abs" é para deixar sempre o resultado em positivo 
  print("Liga Máquina")
else:
  print('Máquina Desligada')
```

```python
## Programa da Senha e Usuario
login_realizado=False
tentativas = 0
while (login_realizado is False and tentativas < 3):
  usuario = input('Insira o seu usuário: ')
  senha = input('Insira a sua senha: ')
  if (usuario.upper() == 'ADMIN' and senha.upper() == '123'):
    print("Você está logado no sistema !")
    login_realizado = True 
  else: 
    print ('Usuário e/ou a Senha está(ão) Incorreta(s)')
    tentativas +=1
print(f'Foram utilizadas {tentativas} tentativas !')
print('Foram utilizadas {} tentativas !'.format(tentativas))

```
